```python
#código 3.1: carregando arquivo Tabela3.1
#faça upload da Tabela3.1 em sample_data (à esquerda)
import pandas as pd
tab31=pd.read_excel('/content/sample_data/Tabela3.1.xlsx')
display(tab31.head(3))
```

```python
##código 3.2: cálculo de SSE, MSE e RMSE
import numpy as np
from sklearn.metrics import mean_squared_error
y=tab31['y']
yhat=tab31['yhat']
# SSE: Soma dos erros ao quadrado
SSE = np.sum((y - yhat) ** 2)

# MSE: Média dos erros ao quadrado
MSE = mean_squared_error(y, yhat)

# RMSE: Raiz quadrada do MSE
RMSE = np.sqrt(MSE)

print(f"SSE: {round(SSE, 2)}")
print(f"MSE: {round(MSE, 2)}")
print(f"RMSE: {round(RMSE, 2)}")
```
